### Step 1. Import the necessary libraries

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Occupation_Exercise").config('spark.sql.ansi.enabled','false').getOrCreate()
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

### Step 3. Assign it to a variable called users and use the 'user_id' as index

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
chipo = pd.read_csv(url,sep='|')

users = spark.createDataFrame(chipo)

```

### Step 4. See the first 25 entries

```python
users.show(25)
```

### Step 5. See the last 10 entries

```python
from pyspark.sql import functions as F
users.orderBy(F.desc('user_id')).show(10)
```

### Step 17. What is the mean age of users?

```python
from pyspark.sql import functions as F
(
    users.agg(
      F.mean("age").alias("age_mean")

    ).withColumn(
        "age_mean",F.round("age_mean",0)
    ).show()
)
```

### Step 18. What is the age with least occurrence?

```python

users.groupBy("age").count().orderBy("count",ascending=True).filter(F.col("count")==1).show()
```
